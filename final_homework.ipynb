Проверить гипотезу, что сотрудники с высоким окладом
проводят на работе больше времени, чем сотрудники с низким
окладом

Среднее количество рабочих часов сотрудников c низким окладом

```python
df1[df1.salary == 'low']['average_montly_hours'].mean()
```

Среднее количество рабочих часов сотрудников c высоким окладом

```python
df1[df1.salary == 'high']['average_montly_hours'].mean()
```

За нулевую гипотезу принимаем, что уровень оклада не оказывает влияния на среднее количество часов, проведенных на работе

Также задаем две переменные salary_low и salary_high

```python
salary_low = df1[df1['salary'] == 'low']['average_montly_hours']
salary_high = df1[df1['salary'] == 'high']['average_montly_hours']
```

Через ttest проводим проверку гипотезы под pvalue

```python
import numpy as np
import scipy.stats as stats
```

```python
result = stats.ttest_ind(salary_low, salary_high, equal_var=False)

a = 0.05

if (result.pvalue < a):
    print('Отвергаем гипотезу H0. В среднем проведённое на работе время отличается в зависимости от уровня зарплаты.')
else:
    print('Принимаем гипотезу H0. В среднем проведённое на работе время одинаково для всех работников.')
```

Рассчитать следующие показатели среди уволившихся и не
уволившихся сотрудников (по отдельности):

*   Доля сотрудников с повышением за последние 5 лет
*   Средняя степень удовлетворенности
*   Среднее количество проектов





Два массива данных: уволившиеся (left=1), работающие (left=0)

```python
actual = df[(df['left'] == 0)]
fired =  df[(df['left'] == 1)]
```

Формируем сводную таблицу из 3-х строк по заданным условиям.
Столбца 2: уволенные и действующие

```python
massive = pd.DataFrame()
rows = []

rows.append(
    {
    'Показатель': 'Доля получивших повышение за 5 лет',
    'Действующие': round(actual.promotion_last_5years.value_counts(normalize=True)[1], 3),
    'Уволившиеся': round(fired.promotion_last_5years.value_counts(normalize=True)[1], 3)
    }
)

rows.append(
    {
    'Показатель': 'Средняя степень удовлетворенности',
    'Действующие': round(actual.satisfaction_level.mean(), 3),
    'Уволившиеся': round(fired.satisfaction_level.mean(), 3)
    }
)

rows.append(
    {
    'Показатель': 'Среднее количество проектов',
    'Действующие': round(actual.number_project.mean(), 3),
    'Уволившиеся': round(fired.number_project.mean(), 3)
    }
)


for row in rows:
    massive = pd.concat([massive, pd.DataFrame([row])])

massive
```

Разделить данные на тестовую и обучающую выборки
Построить модель LDA, предсказывающую уволился ли
сотрудник на основе имеющихся факторов (кроме department и
salary)
Оценить качество модели на тестовой выборки

```python
X = df1.iloc[:,[0,1,2,3,4,5,7]]
y = df1['left']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
model = LinearDiscriminantAnalysis()
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test).round(2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, model.predict(X_test)).round(2)
```
